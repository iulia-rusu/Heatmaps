```python
#non Nan lengths
def non_nan_length(row):
    non_nan_values = row.dropna()
    return len(non_nan_values)
```

```python
lengths_without_Nan = df.apply(non_nan_length, axis=1)
```

```python
#make into a dataframe to graph it
#make a histogram to check the data distribution for better heatmap binning
lengths_without_Nan_data = pd.DataFrame(lengths_without_Nan, columns=['Lengths'])
```

```python
df.set_index("ProteinSubunit",inplace=True)
```

```python
df.iloc[0:2]
```

```python
# Adding a new column that stores the count of non-NaN values in each row
df['NonNaNCount'] = df.apply(non_nan_length, axis=1)
```

```python
df
```

```python
# Using boolean indexing to select rows based on the count of non-NaN values
small_subunits = df[df['NonNaNCount'] < 300]

# Dropping the 'NonNaNCount' column if you don't want it in the result
small_subunits = small_subunits.drop(columns=['NonNaNCount'])
```

```python
small_subunits
```

```python
#defining the rest of the subunit categories by length segregation
medium_subunits = df[(df['NonNaNCount'] >= 301) & (df['NonNaNCount'] <= 500)]
medium_subunits = medium_subunits.drop(columns=['NonNaNCount'])
```

```python
medium_subunits
```

```python
big_subunits = df[(df['NonNaNCount'] >= 501) & (df['NonNaNCount'] <= 800)]
big_subunits = big_subunits.drop(columns=['NonNaNCount'])
```

```python
big_subunits
```

```python
largest_subunits = df[df['NonNaNCount'] > 801]
largest_subunits = largest_subunits.drop(columns=['NonNaNCount'])
```

```python
largest_subunits
```

```python
# plt.figure(figsize=(100, 50))
# plt.figure(figsize=(100, 50))
nperplot = 5
for i, (row_ind, row) in enumerate(small_subunits.iterrows()):
    vals = row.values
    vals = vals[:len(vals) - np.isnan(vals).sum()]
    normed_xvals = np.linspace(0, 1, len(vals))
    plt.plot(normed_xvals, vals, label=row.name, alpha=0.5)
    if i % nperplot == 0 and i != 0:
        plt.legend()
        plt.xlabel("Percentage along protein")
        plt.ylabel("Disorder score")
        plt.show()

plt.legend()
plt.xlabel("Percentage along protein")
plt.ylabel("Disorder score")
plt.show()
```

```python
plt.figure(figsize=(40, 10))

def remove_nan_vals(row):
    vals = row.values
    cleaned_vals = vals[~np.isnan(vals)]
    return pd.Series(cleaned_vals)
   

new_df1 = medium_subunits.apply(remove_nan_vals, axis=1)
plt.yticks(range(len(new_df1)), list(new_df1.index))
plt.xlabel("Amino Acid Position", fontsize=50)
plt.ylabel("Disorder score", fontsize=50)
plt.yticks(fontsize= 20)
plt.xticks(fontsize= 20)
im = plt.imshow(new_df1, cmap="rocket_r", vmax = 1, vmin = 0, aspect='auto' )
plt.colorbar(im,fraction=0.025, pad=0.04)
plt.show()
```
