```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/Users/iuliarusu/Documents/humanSAGA_IDR.csv')
```

```python
df.set_index("ProteinSubunit",inplace=True)
```

```python
df.iloc[0:2]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# plt.figure(figsize=(100, 50))
nperplot = 5
for i, (row_ind, row) in enumerate(df.iterrows()):
    vals = row.values
    vals = vals[:len(vals)-np.isnan(vals).sum()]
    normed_xvals = np.linspace(0,1,len(vals))
    plt.plot(normed_xvals, vals, label=row.name, alpha=0.5)
    if i%nperplot==0 and i!=0:
        plt.legend()
        plt.xlabel("Percentage along protein")
        plt.ylabel("Disorder score")
        plt.show()
        
plt.legend()
plt.xlabel("Percentage along protein")
plt.ylabel("Disorder score")
plt.show()
```

```python
npixels = 50
perc_interval = npixels // 10
plt.figure(figsize=(20, 30))


def chunk_vals(row):
    vals = row.values
    vals = vals[:len(vals)-np.isnan(vals).sum()]
    chunked_vals = np.array_split(vals, npixels)
    return [np.mean(chunk) for chunk in chunked_vals]

chunked_lists = df.apply(chunk_vals, axis=1)

res = np.stack(chunked_lists.values)


plt.imshow(res, cmap="viridis", vmax=1, vmin=0)
plt.yticks(range(len(df)), list(df.index))

all_nones  = [None]*npixels
for i in range(npixels):
    if i%perc_interval==0:
        all_nones[i] = round(i/npixels,2)
all_nones[-1] = 1.0
plt.xticks(range(0,npixels), all_nones)

plt.xlabel("Percentage along protein")
plt.ylabel("Disorder score")
im = plt.imshow(res, cmap="viridis", vmax=1, vmin=0)
plt.colorbar(im,fraction=0.015, pad=0.04)
plt.show()
```
